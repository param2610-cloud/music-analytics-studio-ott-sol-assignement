## Checklist

```python
# 1. Are revenue columns numeric? If not, coerce.
def check_numeric_revenue(df, col_name, dataset_name):
    if not pd.api.types.is_numeric_dtype(df[col_name]):
        print(f"[{dataset_name}] '{col_name}' is NOT numeric. Sample values:")
        print(df[col_name].unique()[:5])
    else:
        print(f"[{dataset_name}] '{col_name}' is numeric.")

check_numeric_revenue(airtel, 'Rev', 'airtel')
check_numeric_revenue(jio, 'income', 'jio')
check_numeric_revenue(wynk, 'income', 'wynk')
```

```python
# 2. if activity period is present, convert to datetime.
# Note: Based on file inspection, we need to identify date columns.
# Airtel doesn't seem to have a clear date column in the first few lines.
# Jio and Wynk might have them or they might be inferred from filenames or other metadata.

print("Checking for potential date columns...")
for name, df in [('airtel', airtel), ('jio', jio), ('wynk', wynk)]:
    print(f"[{name}] Columns: {df.columns.tolist()}")
```

```python
# 3. checking if store_name, artist, album, track, country, unit_type/stream_type present?
required_concepts = ['artist', 'album', 'track', 'country', 'stream_type']

for name, df in [('airtel', airtel), ('jio', jio), ('wynk', wynk)]:
    print(f"--- {name.upper()} Column Check ---")
    for concept in required_concepts:
        # Simple keyword match
        matches = [c for c in df.columns if concept in c.lower()]
        print(f"{concept}: {matches if matches else 'MISSING'}")
    print("\n")
```

```python
# 4. Are there duplicate rows?
for name, df in [('airtel', airtel), ('jio', jio), ('wynk', wynk)]:
    dupes = df.duplicated().sum()
    print(f"[{name}] Duplicate rows: {dupes}")
```

```python
# 5. Any negative values in revenue or stream_count?
def check_negatives(df, col_name, dataset_name):
    if col_name in df.columns and pd.api.types.is_numeric_dtype(df[col_name]):
        neg_count = (df[col_name] < 0).sum()
        print(f"[{dataset_name}] Negative values in '{col_name}': {neg_count}")
    elif col_name in df.columns:
         print(f"[{dataset_name}] '{col_name}' is not numeric, skipping negative check.")

check_negatives(airtel, 'Rev', 'airtel')
check_negatives(jio, 'income', 'jio')
check_negatives(wynk, 'income', 'wynk')

# Check stream counts if columns exist (guessing names based on typical schemas)
stream_cols = ['Stream Count', 'total', 'Total Streams']
for name, df in [('airtel', airtel), ('jio', jio), ('wynk', wynk)]:
    for col in stream_cols:
        if col in df.columns:
             check_negatives(df, col, name)
```
